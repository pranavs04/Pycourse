```python
import pandas as pd 
# importing the data
log_data = pd.read_csv('C:\\Users\\PRANAV SAXENA\\Desktop\\Python\\basic_course-master\\Lecture_3\\data\\log_data_csv.csv')
log_data
```

```python
# E1
import matplotlib.pyplot as plt # importing the library
plt.figure(figsize=(10,12))      # determining the figure size
plt.title('NPHI vs DEPTH', fontsize=18) # assigning plot title
plt.xlabel('NPHI', fontsize=12)  # assigning x-axis title
plt.ylabel('DEPTH', fontsize=12) # assigning y-axis title
plt.scatter(x=log_data['NPHI'], y=log_data['DEPTH'], c= log_data['GR']) # plotting the graph with Neutron porosity on x and depth on y axis and adding color refering to GR
plt.gca().invert_yaxis() # inverting the y-axis
GR_colorbar=plt.colorbar() # adding colorbar for reference
GR_colorbar.ax.set_title('Gamma Ray') # assigning colorbar title
```

```python
# E2
import numpy as np
log_data['LITHOLOGY'] = np.where(log_data.GR>=75, 'SHALE', 'SANDSTONE')
log_data
```

```python
# E2
import matplotlib.pyplot as plt # importing the library
# determining the figure size
plt.title('NPHI vs RHOB', fontsize=18) # assigning plot title
plt.xlabel('NPHI', fontsize=12)  # assigning x-axis title
plt.ylabel('RHOB', fontsize=12) # assigning y-axis title
color = ['red' if i=='SHALE'  else 'blue' for i in log_data['LITHOLOGY']]
plt.scatter(x=log_data['NPHI'], y=log_data['RHOB'], c= color) # plotting the graph with Neutron porosity on x and depth on y axis and adding color refering to GR
```
