```python
# E1 

import numpy as np

arr = np.array([range(0,10)]) # Here using range(x,y) lists numbers that comprises of values in range x to y with x included and y excluded and np.array makes an array of these values
arr
```

```python
# E2
arr_1 = arr % 2==1 # this condition creates a list with boolean values, i.e. , True or False depending on the condition and values

arr_filter = arr[arr_1] # this statement creates the array,arr_filter, with those elements where the condition in previous statement is True
# for eg, let say at 1, the condition holds True while at 0, it holds false. So 1 will be included in arr_filtered while 0 won't.

arr_filter
```

```python
# E3
arr_2 = np.where(arr%2!=0, -1, arr)
# where() function is used to replace the values in an array.
# np.where(condition, new value, array): Condition is checked in the array. New value is the value that will be entered everywhere the condition holds true. Array is that array where changes are to be made.
arr_2
```

```python
# E13
log_data['GR'].describe() # Describes the statistics of GR, gamma ray column
percentile_25 = log_data['GR'].quantile(0.25) # Calculates 25 percentile GR
percentile_75 = log_data['GR'].quantile(0.75) # Calculates 75 percentile GR
min_gr = log_data['GR'].min() # Calculates minimum GR
median = log_data['GR'].median() # Calculates median GR
print ('Minimum GR = '+str(min_gr))
print ('Median GR = '+str(median))
print ('25 percentile gr = '+str(percentile_25))
print ('75 percentile gr = '+str(percentile_75))
```
